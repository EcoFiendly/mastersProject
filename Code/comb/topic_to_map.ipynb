```python
# load df_topic_comp
with open("../../Data/comb_global/df_topic_comp.pkl", "rb") as f:
    df_topic_comp = pickle.load(f)
    f.close()
```

```python
df_topic_comp.columns
```

```python
df_topic_comp
```

```python
df_topic_comp = df_topic_comp.rename(columns={'internalTaxonId':'id_no'})
```

```python
# change probabilities to present/absent
for i in ['Spatial_data', 'Water_pollution_threat', 'Aquatic_ecology',
       'Forest_loss', 'Monitoring_and_rewilding',
       'Range_occurrence_and_occupancy', 'Common_threats', 'Forest_ecosystem',
       'Marine_population_structure', 'Fisheries_threats',
       'Threat_distribution', 'Forest_fragmentation', 'Habitat_loss',
       'Area_based_protection', 'Assessment_criteria', 'Population_dynamics',
       'Conservation_actions']:
    df_topic_comp[i] = df_topic_comp[i].mask(df_topic_comp[i] > 0, 1)
```

```python
df_topic_comp
```

```python
df_spat_topic_comp = df_spat.merge(df_topic_comp, on='id_no', how='inner')
```

df_spat_topic_comp['Top_topic'] = df_spat_topic_comp[['Monitor_island_endemics', 'Water_pollution_threat', 'Aquatic_ecology', 'Forest_loss', 'Monitoring_and_rewilding', 'Range', 'Common_threats', 'Forest_ecosystem', 'Population_structure', 'Fisheries_threats', 'Threat_distribution', 'Forest_fragmentation', 'Habitat_loss', 'Area_based_protection', 'Assessment_criteria', 'Population_dynamics', 'Conservation_actions']].idxmax(axis=1)
